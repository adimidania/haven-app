```python

import torch
from transformers import pipeline
from transformers.utils import is_flash_attn_2_available
from contextlib import asynccontextmanager
from fastapi import FastAPI, File, UploadFile,Query
from pydantic import BaseModel
from typing import Any

pipe = None

# Create FastAPI app with lifespan
@asynccontextmanager
async def app_lifespan(app: FastAPI):
    global pipe
    pipe = pipeline(
    "automatic-speech-recognition",
    model="openai/whisper-large-v3", # select checkpoint from https://huggingface.co/openai/whisper-large-v3#model-details
    torch_dtype=torch.float16,
    device="cuda:0", # or mps for Mac devices
    model_kwargs={"attn_implementation": "flash_attention_2"} if is_flash_attn_2_available() else {"attn_implementation": "sdpa"},
    )
    print({"message": "model loaded!"})
    yield
    pipe.clear()

app = FastAPI(lifespan=app_lifespan)



@app.get("/root")
def read_root():
    return {"message": "Hello, this is a FastAPI endpoint hosted on Google Colab!"}

@app.post("/voice_to_text/")
async def voice_to_text(vocal_file: UploadFile = File(...)):
    temp_audio_file = "/content/temp_audio.mp3"
    # Save the file temporarily
    with open(temp_audio_file, "wb") as f:
        contents = await vocal_file.read()
        f.write(contents)
    outputs = pipe(
    temp_audio_file,
    chunk_length_s=30,
    return_timestamps=False,
    )
    return {"status":"done","output":outputs['text']}

class NameRequest(BaseModel):
    name: str

@app.post("/hey/")
def welcome(request: NameRequest):
    return {"message": f"Hello, {request.name}!"}
```

```python
cc.run_app(app=app)
```
